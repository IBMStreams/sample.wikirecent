# ImageAnalysis Tutorial

This suite notebooks in this directory compose real time streaming analysis applications that use live Wiki data. The initial notebook 
demonstrates connecting to a datasource (wikipedia), filtering updates and viewing the data in browser. The
final notebook applies multiple levels of AI models to locate, extract and score live updates at Wiki data.

Each notebook composes, submits and renders the results as a standalone application. There are no 
no code dependencies between notebooks.
